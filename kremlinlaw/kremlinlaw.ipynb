```python
import pandas as pd
import numpy as np
from chart_studio.plotly import plot, iplot
import plotly.graph_objs as go
```

```python
df_laws = pd.read_json('data/alldoc.jsonl', lines=True, dtype={'id': 'int64', 'date_day': 'int32', 'date_month': 'int32', 'date_year': 'int32', 'pgr_id' : 'int32', 'num' : 'int32', 'docid' : 'int32', 'is_secret' : 'int32'})
```

Подсчитаем пробелы в номерах опубликованных НПА (указах Президента РФ) по годам

```python
print('Пробелы в номерах опубликованных НПА (указах Президента РФ) по годам:')
for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]:
    for docid in [7,]:
        max_num = df_laws.loc[df_laws['date_year'] == year].loc[df_laws['docid'] == 7]['num'].max()
        n_count = df_laws.loc[df_laws['date_year'] == year].loc[df_laws['docid'] == 7]['num'].count()
        print(' - year %d, docid %d: max_num = %d, count = %d, gaps = %d, gaps_percent = %0.2f' % (year, docid, max_num, n_count, max_num - n_count, (max_num - n_count) *100.0 / max_num))
```

```python
df_laws.loc[df_laws['date_year'] == 2018].loc[df_laws['docid'] == 7]['num'].count()
```

```python
group_laws_count = df_laws.loc[df_laws['date_year'] == 2018].loc[df_laws['docid'] == 12].groupby(['date_year', 'date_month', 'date_day']).size()#.nunique('id')[['id']]
#dir(group)
df_laws_count = pd.DataFrame(group_laws_count)
df_laws_count = df_laws_count.sort_values(by=[0], ascending=False)
#df_laws_count
print('year\tmonth\tday\tcount')
for ind, row in df_laws_count.iterrows():
    print('%d\t%d\t%d\t%d' %(int(row.name[0]), int(row.name[1]), int(row.name[2]), row[0]))
```
